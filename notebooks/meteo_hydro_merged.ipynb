```python
def find_meteo_region(row: str) -> str:

    meteo_dir = "../data/KazHydroMet/meteo"
    for region in os.listdir(meteo_dir):
        for station in os.listdir(os.path.join(meteo_dir, region)):
            if row.lower() in station.lower():
                return region
    
    return None


def find_hydro_basin(row: str) -> str:

    meteo_dir = "../data/KazHydroMet/hydro"
    for region in os.listdir(meteo_dir):
        for station in os.listdir(os.path.join(meteo_dir, region)):
            if str(row) in station.lower():
                return region
    
    return None
        
        
```

```python
df['meteo_region'] = df["meteo_station"].apply(find_meteo_region)
df['hydro_region'] = df["basin"].apply(find_hydro_basin)
df.head()
```
